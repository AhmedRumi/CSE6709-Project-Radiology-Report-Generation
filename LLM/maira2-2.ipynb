```python
from transformers import AutoModelForCausalLM, AutoProcessor
from pathlib import Path
import torch

model = AutoModelForCausalLM.from_pretrained("microsoft/maira-2", trust_remote_code=True)
processor = AutoProcessor.from_pretrained("microsoft/maira-2", trust_remote_code=True)

device = torch.device("cuda")
model = model.eval()
model = model.to(device)
```

```python
from PIL import Image
import os

def get_sample_data(frontal_image_path: str, lateral_image_path: str, indication_file_path: str) -> dict[str, Image.Image | str | None]:
    """
    Load chest X-rays from local file paths and read indication from a text file.
    """
    def load_image(image_path: str) -> Image.Image | None:
        """
        Open an image from a given file path or return None if the file does not exist or cannot be opened.
        """
        if not os.path.exists(image_path):
            print(f"File not found: {image_path}")
            return None
        try:
            return Image.open(image_path)
        except Exception as e:
            print(f"Failed to open image at {image_path}: {e}")
            return None

    def read_indication_from_file(file_path: str) -> str:
        """
        Reads the indication text from the given file.
        Returns "None" if the file does not exist or is empty.
        """
        if not os.path.exists(file_path):
            print(f"File not found: {file_path}")
            return "None"
        try:
            with open(file_path, 'r') as file:
                content = file.read().strip()
                return content if content else "None"
        except Exception as e:
            print(f"Failed to read file {file_path}: {e}")
            return "None"

    # Load images
    frontal_image = load_image(frontal_image_path)
    lateral_image = load_image(lateral_image_path)

    # Load indication text
    indication = read_indication_from_file(indication_file_path)

    sample_data = {
        "frontal": frontal_image,
        "lateral": lateral_image,
        "indication": indication,
        "comparison": "None.",
        "technique": "PA and lateral views of the chest.",
        "phrase": "Pleural effusion."  # For the phrase grounding example. This patient has pleural effusion.
    }
    return sample_data
```

```python
import os

all_predictions = []

for i in range(1, 501):
    frontal_image_path = f"/content/frontal_images/frontal_images/{i}.png"
    lateral_image_path = f"/content/lateral_images/lateral_images/{i}.png"
    indication_file_path = f"/content/gen_findings/gen_findings_new/{i}.txt"  # Construct indication file path

    # Check if image files exist before processing
    if not os.path.exists(frontal_image_path):
        print(f"Skipping {frontal_image_path}: File not found.")
        continue

    if not os.path.exists(lateral_image_path):
        print(f"Skipping {lateral_image_path}: File not found.")
        continue

    # Check if indication file exists
    if not os.path.exists(indication_file_path):
        print(f"Skipping {indication_file_path}: File not found.")
        continue

    try:
        sample_data = get_sample_data(
            frontal_image_path=frontal_image_path,
            lateral_image_path=lateral_image_path,
            indication_file_path=indication_file_path,  # Pass the indication file path
        )

        processed_inputs = processor.format_and_preprocess_reporting_input(
            current_frontal=sample_data["frontal"],
            current_lateral=sample_data["lateral"],
            prior_frontal=None,  # Our example has no prior
            indication=sample_data["indication"],
            technique=sample_data["technique"],
            comparison=sample_data["comparison"],
            prior_report=None,  # Our example has no prior
            return_tensors="pt",
            get_grounding=False,  # For this example we generate a non-grounded report
        )

        processed_inputs = processed_inputs.to(device)

        with torch.no_grad():
            output_decoding = model.generate(
                **processed_inputs,
                max_new_tokens=300,  # Set to 450 for grounded reporting
                use_cache=True,
            )

        prompt_length = processed_inputs["input_ids"].shape[-1]
        decoded_text = processor.decode(output_decoding[0][prompt_length:], skip_special_tokens=True)
        decoded_text = decoded_text.lstrip()  # Remove leading space
        prediction = processor.convert_output_to_plaintext_or_grounded_sequence(decoded_text)

        all_predictions.append(prediction)

    except Exception as e:
        print(f"Error processing image pair {i}: {e}")
        continue

print("Processing completed!")
```

```python
print(len(all_predictions))
# Print out predictions

for idx, prediction in enumerate(all_predictions):
    print(f"Prediction for case {idx + 1}:")  # Output case number
    print(prediction)  # Print prediction
    print("-" * 50)  # Separator between cases
```

```python
# Specify the file path to write predictions to
output_file_path = "/content/predictions1_500_with_findings.txt"  # Change this to the desired path where you want to save the file

# Open the file in write mode (it will overwrite the file if it exists)
with open(output_file_path, 'w') as file:
    # Iterate over the predictions and write them into the text file
    for idx, prediction in enumerate(all_predictions):
        file.write(f"Prediction for case {idx + 1}:\n")  # Write case number
        file.write(f"{prediction}\n")  # Write the actual prediction
        file.write("=" * 50 + "\n")  # Add a separator line for readability

    # Optionally, print the total number of predictions written to the file
    print(f"Predictions for {len(all_predictions)} cases have been written to {output_file_path}")
```
