```python
import datasets
import torch
from torch.utils.data import DataLoader
import transformers
from torchvision.transforms import v2
from huggingface_hub import notebook_login
from huggingface_hub import HfApi

# For displaying:
import numpy as np
import matplotlib.pyplot as plt
```

```python
torch.__version__, transformers.__version__
```

```python
mbatch_size = 1
```

```python
# Device:
if torch.backends.mps.is_available():
    device = torch.device('mps')
elif torch.cuda.is_available():
    device = torch.device('cuda')
else:
    device = torch.device('cpu')
```

```python
# Load model:
tokenizer = transformers.AutoTokenizer.from_pretrained('aehrc/cxrmate-rrg24')
model = transformers.AutoModel.from_pretrained('aehrc/cxrmate-rrg24', trust_remote_code=True).to(device=device)
model.eval()

transforms = v2.Compose(
    [
        v2.PILToTensor(),
        v2.Grayscale(num_output_channels=3),
        v2.Resize(size=model.config.encoder.image_size, antialias=True),
        v2.CenterCrop(size=[model.config.encoder.image_size]*2),
        v2.ToDtype(torch.float32, scale=True),
        v2.Normalize(mean=model.config.encoder.image_mean, std=model.config.encoder.image_std),
    ]
)
```

```python
# HF token for accessing the dataset:
api = HfApi()
try:
    user_info = api.whoami()
    if not user_info:
        print("No user token found. Please log in.")
        notebook_login()
        user_info = api.whoami()
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    notebook_login()
```

```python
# Data:
dataset = datasets.load_dataset('StanfordAIMI/interpret-cxr-test-public', token=True)['test']

def transform_batch(batch):
    batch['images'] = [torch.stack([transforms(j) for j in i]) for i in batch['images']]
    batch['images'] = torch.nn.utils.rnn.pad_sequence(batch['images'], batch_first=True, padding_value=0.0)
    return batch

dataset = dataset.with_transform(transform_batch)
dataloader = DataLoader(dataset, batch_size=mbatch_size, shuffle=True)
batch = next(iter(dataloader))
```

```python
# Display images:
for i, j in enumerate(batch['images']):
    display(f'Study {i+1}:')
    for k in j:
        if not torch.all(k == 0):
            image = k.numpy()
            for channel in range(image.shape[0]):
                image[channel] -= image[channel].min()
                image[channel] /= image[channel].max()
            image = (image * 255).astype(np.uint8)
            plt.imshow(image.transpose(1, 2, 0), cmap='gray')
            plt.axis('off')
            plt.show()
```

```python
frontal_folder="/content/drive/MyDrive/NLMCXR/frontal_images/"
lateral_folder="/content/drive/MyDrive/NLMCXR/lateral_images/"
findings_folder="/content/drive/MyDrive/NLMCXR/findings/"

import os
from PIL import Image
import torch
from torchvision import transforms as T
from torch.nn.utils.rnn import pad_sequence

transform = T.Compose([
    T.Resize((384, 384)),  # Resize to a fixed size
    T.ToTensor(),          # Convert to tensor (scales to [0, 1])
])

def is_file_empty(file_path):
    try:
        with open(file_path, 'r') as file:
            content = file.read().strip()
            return len(content) == 0  # True if empty
    except Exception as e:
        print(f"Error reading file {file_path}: {e}")
        return True  # Treat as empty if unreadable

numbers=[]
predictions=[]
findings_actual=[]
# Check the first 600 cases
for i in range(1, 601):
    img_name = f"{i}.png"
    file_name = f"{i}_findings.txt"

    # Construct file paths
    frontal_path = os.path.join(frontal_folder, img_name)
    lateral_path = os.path.join(lateral_folder, img_name)
    findings_path = os.path.join(findings_folder, file_name)

    if os.path.exists(frontal_path) and os.path.exists(lateral_path) and os.path.exists(findings_path):
      if is_file_empty(findings_path):
        continue
      numbers.append(i)
      image1 = Image.open(frontal_path).convert("RGB")  # Convert to RGB if not already
      image2 = Image.open(lateral_path).convert("RGB")

      image1_tensor = transform(image1)  # Shape: [3, 128, 128]
      image2_tensor = transform(image2)  # Shape: [3, 128, 128]
      study_tensor = torch.stack([image1_tensor, image2_tensor])  # Shape: [2, 3, 128, 128]
      batch_images = study_tensor.unsqueeze(0)  # Shape: [1, 2, 3, 128, 128]
      batch = {'images': batch_images}

      output_ids = model.generate(
          pixel_values=batch['images'].to(device=device),
          max_length=512,
          num_beams=4,
          # do_sample=True,
          # temperature=1.2,
          bad_words_ids=[[tokenizer.convert_tokens_to_ids('[NF]')]],
          eos_token_id=tokenizer.sep_token_id
      )
      findings, impression = model.split_and_decode_sections(output_ids, tokenizer)
      # _ = [print(f'Findings:\n\t{i}\n') for i in findings]
      print(i)
      # print(findings[0])
      predictions.append(findings[0])
      with open(findings_path, 'r') as file:
            content = file.read().strip()
      # print(content)
      findings_actual.append(content)


print(len(predictions))
print(len(findings_actual))
# print(numbers)
```

```python

from PIL import Image
import torch
from torchvision import transforms as T
from torch.nn.utils.rnn import pad_sequence

# Step 1: Define the transformation pipeline
transform = T.Compose([
    T.Resize((384, 384)),  # Resize to a fixed size
    T.ToTensor(),          # Convert to tensor (scales to [0, 1])
])

# Step 2: Load the two PNG images (same study)
image1 = Image.open("/content/CXR2_IM-0652-1001.png").convert("RGB")  # Convert to RGB if not already
image2 = Image.open("/content/CXR2_IM-0652-2001.png").convert("RGB")

# Step 3: Apply transformations to the images
image1_tensor = transform(image1)  # Shape: [3, 128, 128]
image2_tensor = transform(image2)  # Shape: [3, 128, 128]

# Step 4: Stack the images into a study
study_tensor = torch.stack([image1_tensor, image2_tensor])  # Shape: [2, 3, 128, 128]

# Step 5: Add an extra dimension for batch size (1 study in the batch)
batch_images = study_tensor.unsqueeze(0)  # Shape: [1, 2, 3, 128, 128]

# Step 6: Create the final batch dictionary
batch = {'images': batch_images}

# Step 7: Print or inspect the result
print(f"Batch 'images' shape: {batch['images'].shape}")  # Shape: [1, 2, 3, 128, 128]
print(batch['images'])

```
